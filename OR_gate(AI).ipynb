**Import the Necessary packages**

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

**Data Initialization:**
Create or store your required data into different variables.
The 'training_data' consist an array with values in it, which will be then given to our model for training and then they will be tested on 'target_data' and once it's done we can evaluate is by using 'evaluate_data'.

```python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]], "float32")
target_data = np.array([[0], [1], [1], [1]], "float32")

evaluate_data = np.array([[0,1],[1,1],[0,0],[1,0],[1,0]], "float32")
```

Below is the code in which our model will be created and it will be also compiled.

```python
my_model = tf.keras.models.Sequential()

#tf.keras.layers.Dense(no. of hidden layers, input parameters)

my_model.add(tf.keras.layers.Dense(4, input_dim=2, activation='relu'))
my_model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

#Compiling the model
my_model.compile(loss='mean_squared_error', optimizer='adam', metrics=['binary_accuracy'])
my_model.summary()
```

Here we train our model with the given data and test it on target data.

```python
training = my_model.fit(training_data, target_data, epochs=500, verbose=2)
```

Here we predict our models compatibility.

```python
print(my_model.predict(evaluate_data).round())
```

Just some plots to visualize our models accuracy and losses over instances.

```python
loss_curve = training.history['loss']
acc_curve = training.history['binary_accuracy']
plt.plot(loss_curve, label='Training')
plt.legend(loc='lower left')
plt.title('Loss')
plt.show()
```

```python
plt.plot(acc_curve, label='Training')
plt.legend(loc='upper right')
plt.title('Accuracy')
plt.show()
```
